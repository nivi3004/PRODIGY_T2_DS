```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the automobile dataset
auto_df = pd.read_csv('Automobile_data.csv')

# Display the first few rows of the dataset to understand its structure
print(auto_df.head())

# Check for missing values
print(auto_df.isnull().sum())

# Check for duplicates
print(auto_df.duplicated().sum())

# Data cleaning:
# 1. Handle missing values
auto_df.dropna(inplace=True) # Drop rows with missing values

# 2. Remove duplicates
auto_df.drop_duplicates(inplace=True)

# 3. Convert data types if needed
# Convert price column to numeric
auto_df['price'] = pd.to_numeric(auto_df['price'], errors='coerce')

# Summary statistics after data cleaning
print(auto_df.describe())

# EDA:

# Distribution of numerical variables
plt.figure(figsize=(12, 8))
sns.histplot(auto_df['price'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='body-style', y='price', data=auto_df, palette='viridis')
plt.title('Price Distribution by Body Style')
plt.xlabel('Body Style')
plt.ylabel('Price')
plt.show()

# Relationship between numerical variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='engine-size', y='horsepower', data=auto_df, hue='fuel-type', palette='Set2')
plt.title('Engine Size vs Horsepower by Fuel Type')
plt.xlabel('Engine Size')
plt.ylabel('Horsepower')
plt.legend(title='Fuel Type')
plt.show()

# Distribution of categorical variables
plt.figure(figsize=(10, 6))
sns.countplot(x='num-of-doors', data=auto_df, palette='pastel')
plt.title('Distribution of Number of Doors')
plt.xlabel('Number of Doors')
plt.ylabel('Count')
plt.show()

# Relationship between numerical and categorical variables
plt.figure(figsize=(10, 6))
sns.boxplot(x='drive-wheels', y='price', data=auto_df, hue='fuel-type', palette='colorblind')
plt.title('Price Distribution by Drive Wheels and Fuel Type')
plt.xlabel('Drive Wheels')
plt.ylabel('Price')
plt.legend(title='Fuel Type')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(auto_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
